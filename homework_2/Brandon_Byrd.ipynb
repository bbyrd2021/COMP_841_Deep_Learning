```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [0.05, 0.1]
y = [0.01, 0.99]
X = np.tile(x, (1000,1))
Y = np.tile(y, (1000,1))

print(X)
print(Y)
```

```python
seed = 10
np.random.seed(seed=seed)
tf.random.set_seed(seed=seed)
```

```python
model = tf.keras.Sequential()
model.add(keras.Input(shape=(2,), name="input_layer"))
model.add(keras.layers.Dense(2, activation='sigmoid', name="hidden_1"))
model.add(keras.layers.Dense(2, activation='sigmoid', name="hidden_2"))
model.add(keras.layers.Dense(2, activation='sigmoid', name="hidden_3"))
model.add(keras.layers.Dense(2, activation='sigmoid', name="output_layer"))
model.compile(optimizer="adam", loss="mse", metrics=["accuracy"])
history = model.fit(X,Y, epochs=10, batch_size=1, shuffle=True)
```

```python
plt.plot(history.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
```
